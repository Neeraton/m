```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=3000
d0=100
p=4
pt=10*np.log10(p*1000)
print(pt)
pl_d0=-32
x0=10.5
n=4
```

```python
pl1=pl_d0+10*n*np.log10(d/d0)
print(pl1)
```

```python
pr1=pt-pl1
print(pr1)
```

```python
pl2=pl_d0+10*n*np.log10(d/d0)+x0
print(pl2)
```

```python
pr2=pt-pl2
print(pr2)
```

```python
n1=[]
n2=[]
n=np.arange(2,6,1)
for i in n:
    pl1=pl_d0+10*i*np.log10(d/d0)
    pr1=pt-pl1
    n1.append(pr1)
    pl2=pl_d0+10*i*np.log10(d/d0)+x0
    pr2=pt-pl2
    n2.append(pr2)
print(n1) 
print(n2) 
```

```python
plt.plot(n,n1)
plt.plot(n,n2)
for xy in zip(n,n1):
    plt.annotate(' (%i,%.2f) '%xy,xy=xy)
for xy in zip(n,n2):
    plt.annotate('(%i,%.2f) '%xy,xy=xy)                
```
