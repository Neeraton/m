```python
import numpy as np
import matplotlib.pyplot as plt
            
num_subcarrier=64
num_symbol=10
mod_order=4
cp_len=16

data=np.random.randint(0,mod_order,size=(num_symbol,num_subcarrier))

qam=np.exp(1j*np.pi*2*data/mod_order)   
        
ifft=np.fft.ifft(qam,axis=1)

cyclic_prefix=ifft[:,-cp_len:]

ofdm=np.hstack([cyclic_prefix,ifft])

ofdm_signal=ofdm.flatten()
plt.plot(ofdm_signal)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.real(ofdm_signal),label='R')
plt.plot(np.imag(ofdm_signal),label='I')
plt.legend()
```
