```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v=72
vm=(v*1000)/3600
print(vm)
```

```python
c=3e8
f=900e6
print(f)
```

```python
lambda_c=c/f
print(lambda_c)
```

```python
theta=np.array([180,0,60,90])
theta_rad=np.deg2rad(theta)
print(theta_rad)
```

```python
fd=(vm*np.cos(theta_rad)/lambda_c)
print(fd)
```

```python
fr=f+fd
print(fr)
```

```python
print("    Theta(degree)   |   fd    |     fr      ")
for i in range(len(theta)):
    print(f"{theta[i]:10.0f} |{fd[i]:20.2f} | {fr[i]:25.2f} ")
```

```python
plt.plot(theta,fd)
```

```python
plt.plot(fd,fr)
```
