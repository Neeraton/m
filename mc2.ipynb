```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def q(N):
    q=(3*N)**0.5
    return q
```

```python
print(q(3))
```

```python
n1=[]
n2=[]
for i in range(1,5):
    for j in range(1,5):
        N=i*i+j*j+i*j
        n1.append(N)
        r=np.round(q(N),2)
        n2.append(r)
print(n1)
print(n2)
        
```

```python
plt.stem(n1,n2)
for xy in zip(n1,n2):
    plt.annotate('(%i,%.2f)'%xy,xy=xy)
```

```python
def si(i0,n,N):
    q=(3*N)**0.5
    si=((q)**n)/i0
    si_dB=10*np.log10(si)
    return si_dB
```

```python
n3=[]
n4=[]
for i in range(1,5):
    for j in range(1,5):
        N=i*i+j*j+i*j
        n3.append(N)
        r=np.round(si(6,4,N),2)
        n4.append(r)
print(n3)
print(n4)
```

```python
plt.stem(n3,n4)
for xy in zip(n3,n4):
    plt.annotate('(%i,%.2f)'%xy,xy=xy)
                 
```
