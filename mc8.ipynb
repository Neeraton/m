```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

```python
snr_db=np.arange(1,21,2)
print(snr_db)
```

```python
snr=[mt.pow(10,i/10) for i in snr_db]
print(snr)
```

```python
siso,simo,mimo=[],[],[]
for i in snr:
    siso.append(np.log2(1+i))
    simo.append(np.log2(1+i*4))
    mimo.append(2*np.log2(1+i))
    
```

```python
plt.plot(snr_db,siso)
plt.plot(snr_db,simo)
plt.plot(snr_db,mimo)
```

```python
M=np.arange(0,13,2)
```

```python
for j in M:
    mimo1=[]
    for k in snr:
        mimo1.append(j*np.log2(1+k))
    plt.plot(snr_db,mimo1)
```
