```python
def sc(N,TA,CA,TC):
    ca=int(N*CA)
    m=int(TA/ca)
    #print("replicaton",m)
    k=int(TC/N)
    #print("Channel per cell",k)
    sc=int(m*k*N)
    #print("System Capacity",sc)
    return sc
```

```python
n1=[]
n2=[]
n3=[1,3,5,7,9,12,15,17,19,21]
for i in range(1,5):
    for j in range(1,5):
        N=i*i+j*j+i*j
        if N in n3:
            e=N
            n1.append(e)
            r=sc(N,4200,12,1001)
            n2.append(r)
print(n1)
print(n2)
print(n3)
```

```python
plt.stem(n1,n2)
for xy in zip(n1,n2):
    plt.annotate('(%i,%.2f)'%xy,xy=xy)
plt.grid()
```
